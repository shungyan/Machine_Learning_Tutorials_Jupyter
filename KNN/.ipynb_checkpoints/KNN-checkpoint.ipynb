```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
target_names=digits.target_names
target_names.shape
```

```python
target=digits.target
target.shape
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
df['target']=digits.target
df
```

```python
df['targets_name'] =df.target.apply(lambda x: digits.target_names[x])
df
```

target names is equal to target

```python
df=df.drop('targets_name',axis='columns')
df
```

```python
from sklearn.model_selection import train_test_split
x_train, y_train, x_test, y_test= train_test_split(df.drop('target',axis='columns'),df.target,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train,y_train)
```
